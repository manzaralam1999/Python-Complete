```python
# TODO Application by another method

'''
marks = "89"
percen = "90%"
print("Marks: {} - percentage : {} ".format(marks,percen))
'''

'''
Main menu
Add task
Delete task
Terminate Application

'''
space =' '*15
def isinteger(userInput):
    try:
        int(userInput)
        return True
    except ValueError:
        return False
    
def mainmenu(menu,space):
    #space =' '*15
    print("%s --------- Main menu --------- \n"% space)
    for menukey , menuName in menu.items():
        print("{}.{}".format(menukey,menuName))
    usercommand = input("\nEnter command :")
    return usercommand

def addtask(taskdic,space):
    
    
    a=0
    for new in taskdic.keys():
        
        if a < new:
            a = new
    print("%s --------- Adding Task --------- \n"% space)
    taskindex = a +1
   
        
  
            
    taskname = input("TASK #{}".format(taskindex))
    taskdic.update({taskindex : taskname})
    print("Task #{} . Successfully Added\n".format(taskindex))
    

def viewtask(taskdic,space):
    print("%s --------- Viewing Tasks --------- \n"% space)
    if len(taskdic) > 0:
        for taskindex,taskname in taskdic.items():
            print("Task #{} - {}".format(taskindex,taskname))
    else:
        print("No task found\n")
        
        
def deletetask(taskdic,space):
    print("%s --------- Deleting Tasks --------- \n"% space)
    if len(taskdic) > 0:
        for taskindex,taskname in taskdic.items():
            print("Task #{} - {}".format(taskindex,taskname))
        
        taskindex = int(input("Enter Task index eg. 1:"))
        if taskindex in taskdic.keys():
            del taskdic[taskindex]
            print("Task #{} - Successfully Deleted".format(taskindex))
        else:
            print("Invalid TAsk ,Try valid Index")
    
    else:
        print("No Task Found\n")
    
def dltall(taskdic,space):
    print("%s --------- Deleting Tasks --------- \n"% space)
    if len(taskdic) > 0:
        taskdic.clear()
        print("Deleting All task succeassfully!")
    else:
        print("No Task Found\n")

menu = {
    1 : "Add Tasks",
    2 : "View Tasks",
    3 : "Delete Tasks",
    4 : "Terminate App",
    5 : "Delete All"
}

taskdic = {}
while (True):
    usercommand = mainmenu(menu,space)
    if isinteger(usercommand) == True:
        usercommand = int(usercommand)
        #Add task
        
            
        if usercommand > 0 and usercommand <= 5:
            if usercommand == 1:
                addtask(taskdic,space)
            if usercommand == 2:
                viewtask(taskdic,space)
            if usercommand ==3:
                deletetask(taskdic,space)
            if usercommand == 4:
                print("Application Terminated")
                break
            if usercommand == 5:
                dltall(taskdic,space)
        else:
            print("\nInvalid command, Try valid command")
    else:
        print("\nInvalid command,Try valid command")
```
